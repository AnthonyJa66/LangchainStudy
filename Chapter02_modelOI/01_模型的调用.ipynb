方式一——通过获取.ven的参数对baseurl和apikey进行赋值

```python
import os
import dotenv
from langchain_openai import ChatOpenAI

from chapter01_summary.test import response

dotenv.load_dotenv()

###########核心代码############
#llm = OpenAI()#纯文本补全模型
llm = ChatOpenAI(
    model="qwen3.6-flash",
    api_key=os.getenv("DASHSCOPE_API_KEY"),
    base_url=os.getenv("DASHSCOPE_BASE_URL"),

)  #对话模型
str = llm.invoke("写一首关于春天的诗")  # 直接输入字符串
#print(str)
print(str.content)
```

```python
import os
import dotenv
from langchain_openai import ChatOpenAI

dotenv.load_dotenv()
os.environ["OPENAI_API_KEY"] = os.getenv("DASHSCOPE_API_KEY")
os.environ["OPENAI_BASE_URL"] = os.getenv("DASHSCOPE_BASE_URL")

###########核心代码############
#llm = OpenAI()#纯文本补全模型
llm = ChatOpenAI(
    model="qwen3.6-flash",
    temperature=0.7,
    max_tokens=256,
)  #对话模型

str = llm.invoke("写一首关于春天的诗")  # 直接输入字符串

#print(str)
print(str.content)
```

阿里百炼平台调用

通过OpenAI sdk实现，pip install openai

```python
import os
from openai import OpenAI

client = OpenAI(

    api_key=os.getenv("DASHSCOPE_API_KEY"),
    base_url=os.getenv("DASHSCOPE_BASE_URL")
)
completion = client.chat.completions.create(
    model="qwen3.6-flash",  # 此处以 deepseek-r1 为例，可按需更换模型名称。qwen3.6-flash
    #max_tokens=256,
    messages=[
        {'role': 'user', 'content': '中文回答：数字9.9和9.11谁大'}
    ]
)
# 通过reasoning_content字段打印思考过程
print("思考过程：")
print(completion.choices[0].message.reasoning_content)

# 通过content字段打印最终答案
print("最终答案：")
print(completion.choices[0].message.content)
```

通过dashscope实现，可以不写baseurl，pip install dashscope

```python
import os
import dashscope

messages = [
    {'role': 'user', 'content': '你是谁？'}
]
response = dashscope.Generation.call(

    api_key=os.getenv('DASHSCOPE_API_KEY'),
    model="deepseek-r1",  # 此处以 deepseek-r1 为例，可按需更换模型名称。
    messages=messages,
    # result_format参数不可以设置为"text"。
    result_format='message'
)
print("=" * 20 + "思考过程" + "=" * 20)
print(response.output.choices[0].message.reasoning_content)
print("=" * 20 + "最终答案" + "=" * 20)
print(response.output.choices[0].message.content)
```
